```python
import pandas as pd
import plotly.graph_objects as go
from ipywidgets import Output, VBox
from google.colab import output as colab_output

# Group and clean the data
df_grouped = df.groupby([df['pubdate'].dt.strftime('%Y-%m-%d'), 'category_narratives']).size().reset_index(name='count')

# Get the number of unique clusters
unique_clusters = df_grouped['category_narratives'].unique()
num_clusters = len(unique_clusters)
# Create a list of unique pubdates
unique_pubdates = df_grouped['pubdate'].unique()

# Initialize a dictionary to hold the data for each cluster
cluster_data = {cluster: [0] * len(unique_pubdates) for cluster in unique_clusters}

# Populate the data
for idx, row in df_grouped.iterrows():
    cluster = row['category_narratives']
    date_idx = list(unique_pubdates).index(row['pubdate'])
    cluster_data[cluster][date_idx] = row['count']

# Convert the cluster data to a 2D array
ls = [go.Bar(name=cluster, x=unique_pubdates, y=cluster_data[cluster], hovertemplate=f'Narrative={cluster}<br>Date=%{{x}}<br>Number of Articles=%{{y}}<extra></extra>') for cluster in unique_clusters]

f = go.FigureWidget(ls)

# Update layout to make the area and charts higher
f.update_layout(
    barmode='stack',
    title='Narrative Change Over Time',
    xaxis_title='Date',
    yaxis_title='Number of Articles',
    legend_title='Narrative',
    height=500  # Increase height of the chart
)

# Create an output widget to display the top N entries
output_widget = Output()

# Create the callback function
def update_bar(trace, points, selector):
    if points.point_inds:
        # Clear the previous output
        with output_widget:
            output_widget.clear_output()

            # Get the clicked data
            idx = points.point_inds[0]
            clicked_date = trace.x[idx]
            clicked_narrative = trace.name

            # Filter the top N (5) entries sorted by df.index desc relevant for 'this group'
            filtered_df = df[(df['pubdate'].dt.strftime('%Y-%m-%d') == clicked_date) & (df['category_narratives'] == clicked_narrative)]
            top_n_df = filtered_df.nlargest(5, 'pubdate').copy().reset_index(drop=True)

            # Display the filtered DataFrame
            display(top_n_df.loc[:, ('Summary', 'Topics', 'Key_People', 'Sentiment', 'Binary_Sentiment', 'Emotions', 'Narratives')])


# Initialize marker.line.width for all bars
for bar in f.data:
    bar.marker.line.width = [0] * len(bar.x)
    bar.marker.line.color = 'black'

# Attach click event to each series
for bar in f.data:
    bar.on_click(update_bar)

# Display the figure and output widget
display(VBox([f, output_widget]))
```

```python
# Assuming df is your DataFrame with the given structure

# Function to normalize the case to sentence case
def normalize_case(s):
    if isinstance(s, list):
        return [i.capitalize() for i in s]
    elif isinstance(s, str):
        return s.capitalize()
    return s

# Normalize the case for Sentiment, Binary_Sentiment, and Emotions
df['Sentiment'] = df['Sentiment'].apply(normalize_case)
df['Binary_Sentiment'] = df['Binary_Sentiment'].apply(normalize_case)
df['Emotions'] = df['Emotions'].apply(normalize_case)

# Extract single value from Binary_Sentiment list
df['Binary_Sentiment'] = df['Binary_Sentiment'].apply(lambda x: x[0] if isinstance(x, list) else x)

# Explode lists to individual rows for aggregation
df_exploded_sentiment = df.explode('Sentiment')
df_exploded_emotions = df.explode('Emotions')

# Aggregate Sentiment counts per category_label
sentiment_counts = df_exploded_sentiment.groupby(['category_label', 'Sentiment']).size().reset_index(name='count')

# Aggregate Binary_Sentiment counts per category_label
binary_sentiment_counts = df.groupby(['category_label', 'Binary_Sentiment']).size().reset_index(name='count')

# Aggregate Emotions counts per category_label
emotion_counts = df_exploded_emotions.groupby(['category_label', 'Emotions']).size().reset_index(name='count')

# Visualize Sentiment per category_label
fig_sentiment = px.bar(sentiment_counts, x='category_label', y='count', color='Sentiment',
                       title='Sentiment Distribution per Category Label',
                       labels={'count': 'Count', 'category_label': 'Category Label', 'Sentiment': 'Sentiment'})

fig_sentiment.update_layout(barmode='stack', xaxis_title='Category Label', yaxis_title='Count')

# Visualize Binary Sentiment per category_label
fig_binary_sentiment = px.bar(binary_sentiment_counts, x='category_label', y='count', color='Binary_Sentiment',
                              title='Binary Sentiment Distribution per Category Label',
                              labels={'count': 'Count', 'category_label': 'Category Label', 'Binary_Sentiment': 'Binary Sentiment'})

fig_binary_sentiment.update_layout(barmode='stack', xaxis_title='Category Label', yaxis_title='Count')

# Visualize Emotions per category_label
fig_emotion = px.bar(emotion_counts, x='category_label', y='count', color='Emotions',
                     title='Emotions Distribution per Category Label',
                     labels={'count': 'Count', 'category_label': 'Category Label', 'Emotions': 'Emotions'})

fig_emotion.update_layout(barmode='stack', xaxis_title='Category Label', yaxis_title='Count')

# Show the figures
fig_sentiment.show()
fig_binary_sentiment.show()
fig_emotion.show()
```

```python

# Assuming df is your DataFrame with the given structure

# Function to normalize the case to sentence case
def normalize_case(s):
    if isinstance(s, list):
        return [i.capitalize() for i in s]
    elif isinstance(s, str):
        return s.capitalize()
    return s

# Normalize the case for Sentiment
df['Sentiment'] = df['Sentiment'].apply(normalize_case)

# Convert pubdate to datetime
df['pubdate'] = pd.to_datetime(df['pubdate'], unit='ms')

# Explode the Sentiment list to individual rows
df_exploded_sentiment = df.explode('Sentiment')

# Aggregate Sentiment counts per date
sentiment_counts_over_time = df_exploded_sentiment.groupby([df_exploded_sentiment['pubdate'].dt.date, 'Sentiment']).size().reset_index(name='count')

# Apply rolling average to smooth the lines
sentiment_counts_over_time['count_smoothed'] = sentiment_counts_over_time.groupby('Sentiment')['count'].transform(lambda x: x.rolling(window=3, min_periods=1).mean())

# Create a line chart to show sentiment change over time
fig_sentiment_over_time = px.line(sentiment_counts_over_time, x='pubdate', y='count_smoothed', color='Sentiment',
                                  title='Smoothed Sentiment Change Over Time',
                                  labels={'pubdate': 'Date', 'count_smoothed': 'Count', 'Sentiment': 'Sentiment'})

fig_sentiment_over_time.update_layout(xaxis_title='Date', yaxis_title='Count')

# Show the figure
fig_sentiment_over_time.show()
```

```python
# Identify top N sentiments
top_n = 5
top_sentiments = sentiment_counts_over_time.groupby('Sentiment')['count'].sum().nlargest(top_n).index

# Filter data to include only top N sentiments
filtered_sentiment_counts = sentiment_counts_over_time[sentiment_counts_over_time['Sentiment'].isin(top_sentiments)]

# Create a line chart to show sentiment change over time
fig_sentiment_over_time_top = px.line(filtered_sentiment_counts, x='pubdate', y='count_smoothed', color='Sentiment',
                                      title=f'Top {top_n} Sentiments Change Over Time',
                                      labels={'pubdate': 'Date', 'count_smoothed': 'Count', 'Sentiment': 'Sentiment'})

fig_sentiment_over_time_top.update_layout(xaxis_title='Date', yaxis_title='Count')

# Show the figure
fig_sentiment_over_time_top.show()
```

```python
# Define a mapping from detailed sentiments to broader categories
sentiment_mapping = {
    'Negative': 'Negative',
    'Positive': 'Positive',
    'Neutral': 'Neutral',
    # Add more mappings as necessary
}

# Apply the mapping to create a new column with broader categories
df_exploded_sentiment['Broad_Sentiment'] = df_exploded_sentiment['Sentiment'].map(sentiment_mapping).fillna('Other')

# Aggregate Broad Sentiment counts per date
broad_sentiment_counts_over_time = df_exploded_sentiment.groupby([df_exploded_sentiment['pubdate'].dt.date, 'Broad_Sentiment']).size().reset_index(name='count')

# Apply rolling average to smooth the lines
broad_sentiment_counts_over_time['count_smoothed'] = broad_sentiment_counts_over_time.groupby('Broad_Sentiment')['count'].transform(lambda x: x.rolling(window=3, min_periods=1).mean())

# Create a line chart to show broad sentiment change over time
fig_broad_sentiment_over_time = px.line(broad_sentiment_counts_over_time, x='pubdate', y='count_smoothed', color='Broad_Sentiment',
                                        title='Broad Sentiment Change Over Time',
                                        labels={'pubdate': 'Date', 'count_smoothed': 'Count', 'Broad_Sentiment': 'Sentiment'})

fig_broad_sentiment_over_time.update_layout(xaxis_title='Date', yaxis_title='Count')

# Show the figure
fig_broad_sentiment_over_time.show()
```

```python
# Create a stacked area chart to show sentiment change over time
fig_sentiment_area = px.area(filtered_sentiment_counts, x='pubdate', y='count_smoothed', color='Sentiment',
                             title=f'Top {top_n} Sentiments Change Over Time (Stacked Area)',
                             labels={'pubdate': 'Date', 'count_smoothed': 'Count', 'Sentiment': 'Sentiment'})

fig_sentiment_area.update_layout(xaxis_title='Date', yaxis_title='Count')

# Show the figure
fig_sentiment_area.show()
```
